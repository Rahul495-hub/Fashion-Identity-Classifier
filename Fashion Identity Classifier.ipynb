## Clustering (Customer Segments)
Goal: Identify groups of customers based on fashion preferences & shopping habits.

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Select only numerical columns for clustering
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
X = df[numerical_cols]

# Standardize the data (important for K-Means)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

print("✅ Data Prepared for Clustering!")
```

```python
# Use Elbow Method to determine the best number of clusters
inertia = []
K_range = range(1, 11)  # Test clusters from 1 to 10

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

# Plot Elbow Curve
plt.figure(figsize=(8,5))
plt.plot(K_range, inertia, marker='o', linestyle='--')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Inertia')
plt.title('Elbow Method to Determine Optimal K')
plt.show()
```

```python
# Choose optimal K based on the elbow method (example: K=3)
optimal_k = 3  # Change this based on the elbow curve result

# Train K-Means model
kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
df['Cluster'] = kmeans.fit_predict(X_scaled)

print("✅ Customers Segmented into", optimal_k, "Clusters!")
```

```python
# Check how many customers are in each cluster
print(df['Cluster'].value_counts())

# Calculate the average values for each cluster
cluster_summary = df.groupby('Cluster').mean()
print("\n📊 Cluster Summary:\n", cluster_summary)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# Reduce dimensions to 2D using PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Add PCA components to DataFrame
df['PCA1'] = X_pca[:, 0]
df['PCA2'] = X_pca[:, 1]

# Plot clusters
plt.figure(figsize=(8,6))
sns.scatterplot(x=df['PCA1'], y=df['PCA2'], hue=df['Cluster'], palette='viridis', s=100, alpha=0.7)
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.title('Customer Segments (K-Means Clustering)')
plt.legend(title='Cluster')
plt.show()
```

```python
# Get mean values for each cluster
cluster_summary = df.groupby('Cluster').mean()

# Display the summary
print("\n📊 Cluster Summary:\n", cluster_summary)
```

```python
# Display summary statistics of each cluster
cluster_summary = df.groupby('Cluster').mean()
print("\n📊 Cluster Summary:\n", cluster_summary)
```

```python
# Define custom cluster names based on insights
cluster_names = {
    0: "Trendsetters (Fashion-Forward)",  
    1: "Budget-Conscious Shoppers",  
    2: "Luxury Enthusiasts",  
    3: "Casual Buyers",
    4: "Minimalists"
}

# Map clusters to names
df['Customer_Segment'] = df['Cluster'].map(cluster_names)

# Display first few rows with assigned segments
print(df[['Cluster', 'Customer_Segment']].head(10))
```
